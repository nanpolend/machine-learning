```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import the necessary function


# 建立空白影像
gc = np.zeros((300, 300, 3), dtype=np.uint8)
gc[:] = [250, 250, 250]

# 在影像上新增文字
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(gc, 'OpenCV', (10, 120), font, 2, (110, 110, 110), 2, cv2.LINE_AA)

# 顯示影像
cv2_imshow(gc) # Use cv2_imshow instead of cv2.imshow

# 等待並關閉
cv2.waitKey(0)
cv2.destroyAllWindows()
```
