```python
# 導入必要的庫
import pandas as pd  # 用於數據處理和操作
import numpy as np  # 提供數值運算和陣列支持
from sklearn import datasets  # 用於載入 Iris 數據集
import matplotlib.pyplot as plt  # 用於創建可視化圖表

# 載入和準備數據
iris = datasets.load_iris()  # 載入 Iris 數據集
x = pd.DataFrame(iris.data, columns=iris.feature_names)  # 將數據轉換為 DataFrame
x.columns = ["sepal_Length", "sepal_width", "petal_Length", "petal_width"]  # 重新命名 DataFrame 的欄位
target = pd.DataFrame(iris.target, columns=["target"])  # 將目標變數轉換為 DataFrame
y = target["target"]  # 提取目標變數
X = x  # 將 x 賦值給 X，可能是筆誤，應移除或修正

# 創建可視化圖表
colmap = np.array(["r", "g", "y"])  # 定義顏色映射，用於根據目標類別為點著色
plt.figure(figsize=(10, 5))  # 創建一個特定大小的圖表
plt.subplot(1, 2, 1)  # 將圖表劃分為 1 行 2 列，並選擇第一個子圖
plt.subplots_adjust(hspace=.5)  # 調整子圖之間的間距
plt.scatter(X["sepal_Length"], X["sepal_width"], color=colmap[y])  # 創建散點圖，使用萼片長度和寬度
plt.xlabel("萼片長度")  # 添加 x 軸標籤
plt.ylabel("萼片寬度")  # 添加 y 軸標籤
plt.subplot(1, 2, 2)  # 選擇第二個子圖
plt.scatter(x["petal_Length"], x["petal_width"], color=colmap[y])  # 創建散點圖，使用花瓣長度和寬度
plt.xlabel("花瓣長度")  # 添加 x 軸標籤
plt.ylabel("花瓣寬度")  # 添加 y 軸標籤
plt.show()  # 顯示創建的圖表
```
