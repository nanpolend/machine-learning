```python
import cv2  # 匯入 OpenCV 函式庫
import numpy as np  # 匯入 NumPy 函式庫
from google.colab.patches import cv2_imshow  # 匯入 Google Colab 專用顯示函式

# 建立一個 300x300 的空白影像，3 個顏色通道 (RGB)，資料型態為 uint8
gc = np.zeros((300, 300, 3), dtype=np.uint8)

# 設定背景顏色為灰白色 (250, 250, 250)
gc[:] = [250, 250, 250]

# 繪製一條線：
# - 起點：(20, 50)
# - 終點：(200, 200)
# - 顏色：(255, 255, 100) (淡黃色)
# - 粗細：5 像素
cv2.line(gc, (20, 50), (200, 200), (255, 255, 100), 5)

# 顯示影像
cv2_imshow(gc)
```
